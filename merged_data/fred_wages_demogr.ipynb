```python
import pandas as pd

macro = pd.read_csv("../data/raw/fred/wages_productivity_data.csv", parse_dates=["Unnamed: 0"])
macro = macro.rename(columns={"Unnamed: 0": "Date"})
macro.set_index("Date", inplace=True)
print(f"Macro dataset: {macro.shape[0]} rows | {macro.shape[1]} columns")

demo = pd.read_csv("../data/raw/fred/macro_demographs.csv", parse_dates=["Unnamed: 0"])
demo = demo.rename(columns={"Unnamed: 0": "Date"})
demo.set_index("Date", inplace=True)
print(f"Demo dataset: {demo.shape[0]} rows | {demo.shape[1]} columns")

macro_q = macro.resample("Q").mean()
demo_q = demo.resample("Q").mean()

merged = macro_q.join(demo_q, how="inner")

print("\n Merge Completed")
print(f"Merged shape: {merged.shape[0]} rows | {merged.shape[1]} columns")
print("\nColumns:", list(merged.columns))
print("\nDate range:", merged.index.min().date(), "to", merged.index.max().date())

merged.to_csv("../data/merged/fred_merged/wages_demogr_merged.csv")
print("\n Saved merged file to ../data/merged/fred_merged/wages_demogr_merged.csv")

merged.head(10)
```
